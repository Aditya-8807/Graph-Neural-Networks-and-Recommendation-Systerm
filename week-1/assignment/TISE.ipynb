```python
import numpy as np
import matplotlib.pyplot as plt

def wavefunction(x, E, V):
    if E > V:
        # Oscillation
        k = np.sqrt(2 * (E - V))
        return np.cos(k * x)
    elif E < V:
        # Decaying
        kappa = np.sqrt(2 * (V - E))
        return np.exp(-kappa * x)
    else:
        # Threshold
        return np.ones_like(x)
```

```python
x = np.linspace(0, 5, 500)
# Graph
psi_A = wavefunction(x, 12, 2)  # E > V
psi_B = wavefunction(x, 2, 7)   #  E < V
psi_C = wavefunction(x, 5, 5)   # E = V

plt.plot(x, psi_A, label='Case A (Oscillatory): E=12, V=2', color='blue', linestyle='-')
plt.plot(x, psi_B, label='Case B (Decaying): E=2, V=7', color='red', linestyle='--')
plt.plot(x, psi_C, label='Case C (Threshold): E=5, V=5', color='green', linestyle='-.')

plt.xlabel('X')
plt.ylabel('psi_X')
plt.grid(True)
plt.savefig('tise_wavefunctions.png', bbox_inches='tight')
plt.show()
plt.close()
```
